## 설치
### OSX
dmg 바이너리 파일이나 juno를 사용할 수 있지만, 여기서는 atom editor를 이용하여 julia를 사용할 것이므로
Homebrew를 통해 설치합니다.

Homebrew 이용 (HomeBrew 설치: http://brew.sh/index_ko.html)

Stable Version 설치

```
> brew tap staticfloat/julia
> brew install --64bit julia
```

Dev Version 설치

```
> brew tap staticfloat/julia
> brew install --HEAD --64bit julia
```
