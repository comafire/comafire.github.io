### Jupyter, Pelican & Github Pages로 블로깅하기

github 하면 개발 소스 저장소가 가장 먼저 떠오르지만, github 에서는 얼마전부터 pages 라는 서비스를 통해 github 저장소에 올려놓은 웹 페이지 소스를 브라우저를 통해 볼수 있도록하면서, 웹 서버가 없어도 개인이 static 블로깅이 가능하도록 서비스하고 있습니다.

wordpress, blogger, tistory 등등 수많은 블로그 시스템이 존재하지만, 개발을 하다가 그때그때 편리하게 블로깅 가능한 면에서는 이 조합을 따라올 수 있는 블로깅 환경이 없는거 같습니다.

https://github.com/ 에 계정을 만듭니다. 

pelican(http://docs.getpelican.com/en/3.6.3/)으로 블로깅을 하기 위해서 2개의 repository 를 생성합니다.

* [자신의 github id].github.io : 이 계정은 pages 를 위한 저장소입니다.
* [자신의 github id].github.io-src : 이 계정은 pelican 을 통해 pages 에 게시될 글들의 저장하기 위한 저장소 입니다.

github 에서는 맥 사용자를 위해 Gtihub Desktop이라는 편리한 github gui 클라이언트 툴을 제공합니다. (https://desktop.github.com/) 

이제 저장소를 clone 하고 pip 를 이용하여 pelican 을 설치합니다.

```bash
> git clone https://github.com/comafire/comafire.github.io
> git clone https://github.com/comafire/comafire.github.io-src
> pip install pelican markdown
```

이제 src 저장소로 이동하여 pelican 에 ipynb 플러그인을 설치합니다. 

ipynb 플러그인은 jupyter 로 작성한 글을 html 형식으로 변환해 주는 플러그인입니다. ipynb 플러그인을 통해 변환된 html 파일이 pelican 의 기본 테마와 잘 맞지 않아 저는 clean-blog 테마를 좀 수정하여 사용하고 있습니다. 테마 설치도 같이 합니다.

```bash
> cd comafire.github.io-src
> pelican-quickstart
> git submodule add git://github.com/danielfrg/pelican-ipynb.git plugins/ipynb
> git submodule add git://github.com/gilsondev/pelican-clean-blog themes/clean-blog
> ls
LICENSE			develop_server.sh	output			plugins			workspace
Makefile		fabfile.py		pelicanconf.py		publishconf.py
content			footer			pelicanconf.pyc		themes
> ls plugins/
ipynb
> ls themes/
clean-blog
```

이제 ipynb 플러그인 및 테마 사용을 위하여 pelicanconf.py 설정파일을 수정해줍니다.

```bash
> vi pelicanconf.py
...
THEME = "/Users/comafire/Projects/comafire.github.io-src/themes/clean-blog"

MARKUP = ('md', 'ipynb')

PLUGIN_PATH = './plugins'
PLUGINS = ['ipynb.markup']
...
```

이제 jupyter 로 작성한 ipynb 노트북 파일을 meta 파일과 함께 content 디렉토리안에 위치 시킵니다.

```bash
> cd content
> ls
2016-08-26-Jupyter-From-Dev-To-Blogging.ipynb
2016-08-26-Jupyter-From-Dev-To-Blogging.ipynb-meta
```

meta 파일의 내용은 아래와 같이 채워 넣으면 됩니다. 글 작성시마다 파일을 하나더 생성해야 하는 번거로움은 있지만, 참을만 합니다. ㅋ

```
Title: Jupyter로 개발부터 블로깅까지
Slug: 2016-08-26-Jupyter-From-Dev-To-Blogging
Date: 2016-08-26
Category: Jupyter
Tags: Pelican, Github Pages
Author: comafire
```

이제 작성한 글을 pelican 을 통해 html 으로 변환 시킵니다.

```
> pelican content
Done: Processed 1 articles, 0 drafts, 0 pages and 0 hidden pages in 0.46 seconds.
> ls output
2016-08-26-Jupyter-From-Dev-To-Blogging.html			
...
```

output 폴더의 내용을 github pages 를 위해 만들어 놓은 저장소에 복사해놓고 commit & push 하면 끝입니다.

```
> cp -R output/* ../comafire.github.io
> cd comafire.github.io/
> git add .
> git commit -m 'post blogging'
> git push
```

이제 github pages 에 접속하면 말끔하게 생성된 블로그 화면을 볼수 있습니다.

이 방식을 이용하면 블로그 작성 소스들과 글들의 버전 관리도 github를 통해 덩달아 되기 때문에 백업 걱정 없이 블로깅이 가능해집니다.
